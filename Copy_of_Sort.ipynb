```
def merge(D,L,R):
  l = 0
  r = 0
  nl = len(L)
  nr = len(R)
  for i in range(len(D)):
    if l >= nl:
      D[i] = R[r]
      r += 1
    elif r >= nr:
      D[i] = L[l]
      l += 1
    else:
      if L[l] < R[r]:
        D[i] = L[l]
        l+=1
      else:
        D[i] = R[r]
        r +=1

def merge_sort(A):
  if len(A) > 1:
    L = A[:len(A)//2]
    R = A[len(A)//2:]
    merge_sort(L)
    merge_sort(R)
    merge(A,L,R)
  return A
```

```
#Quick sort not in place

def quicksort_not_in_place(A):
  if len(A) > 1:
    p = A[0]
    L = []
    E = [p]
    R = []
    for i in range(1,len(A)):
      if A[i]<p:
        L.append(A[i])
      elif A[i] > p:
        R.append(A[i])
      else:
        E.append(A[i])
    L = quicksort_not_in_place(L)
    R = quicksort_not_in_place(R)
    
    A = L+E+R
  return A
```

```
# Quisck Sort com lumuto

def LumutoPartition(A,l,r):
  p = A[l]
  s = l
  for i in range(l+1,r+1):
    if A[i] < p:
      s+=1
      A[s],A[i] = A[i],A[s]
  A[s],A[l] = A[l],A[s]
  return s

def Hoare(A,i,j):
  l = i
  p = j
  r = j-1
  while l<=r:
    while l<=r and A[l] < A[p]:
      l +=1
    while l<=r and A[r] > A[p]:
      r -=1
    if l<=r:
      A[l],A[r] = A[r],A[l]
      l+=1
      r-=1
  A[l],A[p] = A[p],A[l]
  return l

def quicksort_sec(A,l,r,func):
  if l < r:
    s = func(A,l,r)
    quicksort_sec(A,l,s-1,func)
    quicksort_sec(A,s+1,r,func)
  return A
  

def quicksort_lomuto(A):
  return quicksort_sec(A,0,len(A)-1, LumutoPartition)

  
  
def quicksort_Hoare(A):
  return quicksort_sec(A,0,len(A)-1,Hoare)
```

```
import time
from numpy import random


#scope of the randomness
DOMAIN = 1000 #1B

#defines the number of times each algorithm will be processed to find the average time
num_rounds = 15

alg_results = dict()

#calculates the executions average time
def avgTime(func, size, debug=False):
  t = 0
  for i in range(num_rounds):    
    random.seed(size+i+1)
    #create a list with random numbers
    dataList = list(random.randint(DOMAIN, size=size))
    L = dataList.copy()
    start = time.time()
    func(L)
    end = time.time()    
    t += end - start
      
  return t/num_rounds


#defines the algorithms to be processed
algorithms = [bubbleSort,bubbleSort_opt,insertionSort,merge_sort, quicksort_not_in_place, 
              quicksort_lomuto,quicksort_Hoare,HeapSort_Not_in_place,Heapsort_in_place]
#algorithms = [quicksort_lumen]

#sizes  
sizes = [5000,10000,15000,20000,25000]
mapSizeToTime = dict()
for i in range(len(sizes)):
  print(f"Starting collect {i+1}")
  
  #map list size to algorithm average time
  for algorithm in algorithms:
    print('  > ', algorithm.__name__)
    mapSizeToTime[f"{algorithm.__name__ }##{sizes[i]}"] = avgTime(algorithm, sizes[i], True)
  print("")

print(f"Finish data collection")
```

Formata dados

```
import pandas as pd

df = pd.DataFrame.from_dict(mapSizeToTime, orient='index', columns=['Time'])
df['Algorithm'] = [i.split("##")[0] for i in df.index]
df['Size'] = [int(i.split("##")[1]) for i in df.index]
df
```
